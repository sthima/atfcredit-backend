```python
df = df[['CNPJ','Tradução perda', 'Data de perda']]
```

```python
df_final = dfs.merge(df, left_on = 'SConsCnpjCpf', right_on = 'CNPJ', how = 'left')
```

```python
def compare_data(line):
    d1 = line['SConsData']
    d2 = line['Data de perda']
    
    if d2>d1:
        return abs((d2-d1).days)
    else:
        return abs((d1-d2).days)

df_final['Dif_Days'] = df_final.apply(compare_data, axis = 1)
```

```python
df_final_2 = df_final.groupby('CNPJ')[['Dif_Days']].min().reset_index()
```

```python
df_final_3 = df_final_2.merge(df_final, on = ['CNPJ','Dif_Days'],  how = 'left')
```

```python
df_final_3 = df_final_3.groupby('CNPJ').first()
```
