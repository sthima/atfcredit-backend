```python

import pymongo
import pandas as pd
import json
from datetime import datetime
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
df = pd.read_csv('seras.csv', sep = ";", header=None )
df[0] = df[0].astype(int, errors = 'ignore')
```

```python
def get_txt_file(name):
    f = open("TxtLimpo/"+name, "r")
    return f.read()
```

```python

many_list = []
error_txt = []

for i,j in df[df[8].notnull()].iterrows():
    query = { "cnpj": j[0] };

    try:
        content = get_txt_file(j[8])
    
    except:
        error_txt.append(j[8])
        continue
        
    dict_add = { "cnpj":j[0],
                "txt_type":"LIMPO",
                "txt_content":content,
                "isProcessed":False,
                "insertDate":datetime.now() }
#     update = { "$set": dict_add};
    many_list.append(dict_add)
```

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["test"]
mycol = mydb['initial-collection']
```

```python
mycol.insert_many(many_list)
```
