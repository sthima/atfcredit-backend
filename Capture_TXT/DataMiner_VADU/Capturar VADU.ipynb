```python
from threading import Thread
from IPython.display import display
import time

parar = 0
api = []
crawler = []

def busca():
    global parar
    global api
    global crawler

    aaa = 0

    for i,j in zip(new_df['BUSCA'],new_df['CNPJ']):
        if parar>0:
            return api, crawler
        
        print(aaa,' -> ',i, ' - ', j)
        aaa+=1
        try:
            crawler_result = VaduCrawler(i).capture_data()

            if len(crawler_result):
                cn = re.sub(r'[^\w\s]','',crawler_result['cnpj'].iloc[0])
            else:
                if type(j) != type(0.0):
                    cn = re.sub(r'[^\w\s]','',j)
                else:
                    continue
                    
            vadu_result = VaduApi(cn).capture_data()
            crawler_result['BUSCA'] = i
            crawler.append(crawler_result)
            
            vadu_result['BUSCA'] = i
            api.append(vadu_result)
        except:
            continue
    

busca = Thread(target=busca)
busca.start()
```
