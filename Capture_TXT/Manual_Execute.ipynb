```python
import pandas as pd
import numpy as np
import json
import os

import DataMiner
```

```python
df_cases = pd.read_excel('Cases.xlsx', index_col = 0, header=1).reset_index(drop = True)
```

```python


dbm = DataMiner.DataBaseManager()
for file_name in os.listdir('txtPuro/'):
    print(file_name)
    tfm = DataMiner.TextFileManager('txtPuro/'+file_name)

    objs = tfm.create_tables_dict()
    try:
        objs['Resultado'] = df_cases[df_cases['txt'] == file_name].iloc[0]['Tipo']
    except:
        objs['Resultado'] = 'Indefinido'    
        
    dbm.save_PURE_TXT([objs])
    del(tfm)
    
    print(' ')
```
